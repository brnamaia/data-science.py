```python
import pandas as pd
```

```python
microdadosEnem = pd.read_csv(r"C:\Users\brnam\OneDrive\Área de Trabalho\analise-dados\projetos\enem\microdados_enem2018\DADOS\MICRODADOS_ENEM_2018.csv", sep=";", encoding="ISO-8859-1", nrows = 100000)
microdadosEnem.head()
```

```python
microdadosEnem.columns.values
```

```python
microdadosEnemSelecionado = microdadosEnem.filter(items=['CO_MUNICIPIO_RESIDENCIA', 'NO_MUNICIPIO_RESIDENCIA', 'CO_UF_RESIDENCIA', 'SG_UF_RESIDENCIA', 'NU_NOTA_CN', 'NU_NOTA_CH', 'NU_NOTA_LC',
       'NU_NOTA_MT', 'NU_NOTA_REDACAO'])
```

```python
microdadosEnemSelecionado.head()
```

```python
mediaMunicipio = microdadosEnemSelecionado.groupby(['CO_MUNICIPIO_RESIDENCIA', 'NO_MUNICIPIO_RESIDENCIA', 'CO_UF_RESIDENCIA', 'SG_UF_RESIDENCIA'], as_index = False).mean()
```

```python
mediaMunicipio.head()
```

```python
mediaMunicipio.to_csv("mediaMunicipio.csv")
```
