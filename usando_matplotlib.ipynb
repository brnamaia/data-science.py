```python
import numpy as np
```

```python
x = np.arange(1, 10, 0.1)
figl, f1_axes = plt.subplots(ncols=2, nrows=2, figsize=(15, 10))
figl.suptitle("Primeira figura", size=40)

f1_axes[0, 0].plot(np.sin(x), label="sen x", color="lightcoral")
f1_axes[0, 0].set_title("Caixa 1")
f1_axes[0, 0].plot(np.cos(x), label="cos x", color="teal")
f1_axes[0, 0].set_xlabel("Caixa 1 - Label X")
f1_axes[0, 0].set_ylabel("Caixa 1 - Label Y")
f1_axes[0, 0].legend()


f1_axes[0, 1].plot(np.cos(x), label="cos x", color="black")
f1_axes[0, 1].set_title("Caixa 2")
f1_axes[0, 1].set_xlabel("Caixa 2 - Label X")
f1_axes[0, 1].set_ylabel("Caixa 2 - Label Y")
f1_axes[0, 1].legend()

f1_axes[1, 0].plot(np.tan(x), label="tan x", color="olive")
f1_axes[1, 0].set_title("Caixa 3")
f1_axes[1, 0].set_xlabel("Caixa 3 - Label X")
f1_axes[1, 0].set_ylabel("Caixa 3 - Label Y")
f1_axes[1, 0].legend()

f1_axes[1, 1].plot(np.sin(x), label="sen x", color= "gold")
f1_axes[1, 1].set_title("Caixa 4")
f1_axes[1, 1].set_xlabel("Caixa 4 - Label X")
f1_axes[1, 1].set_ylabel("Caixa 4 - Label Y")
```
